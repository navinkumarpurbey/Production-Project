# EDA - Exploratory Data Analysis

```python
# First Import the all libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import the Warnings to igonore the error while EDA process
# Ignoring warnings can hide important issues, such as deprecated functions or potential code errors.
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the Datasets where we cleaned previsouly
df = pd.read_csv("cleaned_healthcare_market_trends.csv")
```

```python
# Now the checking the basic Info about cleaned datasets
print(df.info())
```

```python
print(df.describe())
```

```python
# This line of code helps display all the columns with specific rows
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
df.head(10)
```

```python
# Extract the year (last 4 digits of the timestamp string)
df['Year'] = df['Year'].astype(str).str[-4:].astype(int)

# Verify
print("Years after cleaning:", df['Year'].unique())
```

```python
# This is Correcting version
print("Years in data:", df['Year'].unique())
print("Prevalence Rates:", df['Prevalence Rate (%)'].unique())
```

```python
# Convert to percentages (0.04 â†’ 4%)
df['Prevalence Rate (%)'] = df['Prevalence Rate (%)'] * 100

# Verify
print("Prevalence Rates after cleaning:", df['Prevalence Rate (%)'].unique())
```

```python
# Visualization the Data Distrubtions
# Ploting the histogram for numricial columns
df.hist(figsize=(14, 10), bins = 20)
plt.show()
```

```python
# Identify the Relationships between Variables (Correleatios Heatmap)
plt.figure(figsize = (12, 9))
sns.heatmap(df.corr(), annot = True, cmap = "coolwarm", fmt = ".2f")
plt.title("Correlations Heatmap")
plt.show()
```

```python
# Every exceution the code first saw the Datatype for clearity 
print(df.dtypes)
```

```python
# Checking process
print(df['Healthcare Access (%)'].describe())
print(df['Year'].unique())
```

```python
# Now Analyze the Trends Over Time using Time-series Analysis
# Healthcare spending overtime/over the years
# Actually in which year is Disease is trending over time
plt.figure(figsize=(12, 6))
sns.lineplot(
    x='Year', 
    y='Healthcare Access (%)',
    data=df, 
    hue='Country', 
    marker='o'
)
plt.title("Healthcare Access Trends Over Time (by Country)")
plt.xlabel("Year")
plt.ylabel("Healthcare Access (%)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Now distributions the HealthCare Services by Regions/Country
plt.figure(figsize = (12, 6))
sns.countplot(data = df, x = "Country", order = df["Country"].value_counts().index)
plt.title("Distrubation of HealthCare Services by the CountryðŸ«¡")
plt.xticks(rotation = 90)
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
# Checking for extreme values in relevant columns
print(df[['Healthcare Access (%)', 'Mortality Rate (%)', 'Population Affected', 'Urbanization Rate (%)']].describe())
```

```python
# Now Analysis the Healthcare spending vs Life Expectancy
# Don't touch the line of code because it takes too much time
sns.set(style="whitegrid")

# Sample 5% of the data for faster plotting
df_sampled = df.sample(frac=0.05, random_state=42)

plt.figure(figsize = (8, 4), dpi = 80)
sns.scatterplot(data = df, x = "Healthcare Access (%)", y = "Mortality Rate (%)", hue = "Disease Category", size = "Population Affected", style = "Urbanization Rate (%)", sizes = (20, 200), palette = 'viridis', legend = 'full')
plt.title("Healthcare Access vs Mortality Rate by Disease Category (2000-2023)", fontsize = 14)
plt.xlabel("Healthcare Access (%)", fontsize = 12)
plt.ylabel("Mortality Rate (%)", fontsize = 12)
plt.tight_layout()
plt.show()
```

```python
# Now Analysis the Healthcare Access vs Recovery Rate
plt.figure(figsize = (12, 6))
sns.scatterplot(data = df, x = "Healthcare Access (%)", y = "Mortality Rate (%)", hue = "Country", size = "Population Affected")
plt.title("Healthcare Access vs Recovery Rate")
plt.grid(True)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Identify Outliers and Trends
# Drawing the Box Plot for identify the Outliers

plt.figure(figsize = (12, 6))
sns.boxplot(data = df["Healthcare Access (%)"])
plt.title("Box plot for Healthcare Access percentage Rate (%): ")
plt.show()
```

```python
# Regional Level Analysis
# Healthcare cost by the top(10) Country
top_countries = df.groupby('Country')["Healthcare Access (%)"].mean().sort_values(ascending = False).head(10)
top_countries.plot(kind = 'bar', figsize = (12, 6), color = 'orange')
plt.title("Top 10 countries by Healthcare Access (%)")
plt.ylabel("Average Healthcare Access (%)")
plt.xlabel("Country")
plt.show()
```

```python
#  Visualize Healthcare Access vs Mortality Rate
plt.figure(figsize=(12, 6))
sns.scatterplot(
    x='Healthcare Access (%)', 
    y='Mortality Rate (%)', 
    data=df, 
    hue='Disease Category',
    palette='Spectral',
    alpha=0.7
)
plt.title('Healthcare Access vs Mortality Rate', fontsize=14, fontweight='bold')
plt.xlabel('Healthcare Access (%)', fontsize=12)
plt.ylabel('Mortality Rate (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Disease Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Group by Disease and calculate total prevalence rate
total_prevalence = df.groupby('Disease Name')['Prevalence Rate (%)'].sum()

# Sort and fetch top 5 diseases based on total prevalence rate
top_5_diseases = total_prevalence.sort_values(ascending=False).head(5)

print("\n Top 5 Most Prevalent Diseases ðŸ˜’ðŸ˜’ðŸ˜’")
print(top_5_diseases)
```

```python
# Exploring Healthcare Access vs Population Affected
plt.figure(figsize=(12, 6))
sns.scatterplot(
    x='Healthcare Access (%)', 
    y='Population Affected', 
    data=df, 
    hue='Disease Category',
    palette='Set1'
)
plt.title('Healthcare Access vs Population Affected', fontsize=14, fontweight='bold')
plt.xlabel('Healthcare Access (%)', fontsize=12)
plt.ylabel('Population Affected', fontsize=12)
plt.legend(title='Disease Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.legend(title='Disease Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Filter data for Prevalence Rate = 20% in 2020
df_prevalent = df[
    (df['Prevalence Rate (%)'] == 20.0) & 
    (df['Year'] == 2020)
]

if not df_prevalent.empty:
    most_prevalent_disease = df_prevalent['Disease Name'].value_counts().idxmax()
    print(f"Most Prevalent Disease in 2020 (20% Prevalence): {most_prevalent_disease}")
else:
    print("No diseases with 20% prevalence in 2020 found.")
```

```python
# Now saving the EDA Performed Data
df.to_csv("cleaned_healthcare_market_trends.csv", index = False)
print("EDA Perform Completed. Cleaned file saved!")
```
